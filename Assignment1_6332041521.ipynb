## How many rows are there in the GBvideos.csv after removing duplications?

```python
df = pd.read_csv('data/GBvideos.csv')
df.drop_duplicates(inplace = True)
df
```

```python
print(str(df.shape[0])+" Rows")
```

## How many VDO that have "dislikes" more than "likes"?  <font color=red>DO NOT group by the data and make sure that you count only unique title!</font>

```python
dfunique = df[(df.dislikes>df.likes)][['title','dislikes','likes']]
dfunique.drop_duplicates(subset='title',inplace=True)
print(str(dfunique.shape[0])+" VDOs")
```

## How many VDO that are trending on 22 Jan 2018 with comments more than 10,000 comments?

```python
condition_list = (df['trending_date'] == '18.22.01') & (df['comment_count']> 10000)
print(str(df[condition_list].shape[0])+" VDOs")
```

```python
#datetime = pd.to_datetime(df.publish_time)
#CommentAverage = df.groupby(pd.DatetimeIndex(df.publish_time).normalize())['comment_count'].mean()
CommentAverage = df.groupby(df.trending_date)['comment_count'].mean()
MinCommentAverage= CommentAverage.idxmin()
print(MinCommentAverage)
```

## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?

```python
import json
with open('data/GB_category_id.json') as GBjson:
  category = json.load(GBjson)
category_list = []
for d in category['items']:
  category_list.append((int(d['id']), d['snippet']['title']))
category_list 
```

```python
category_df = pd.DataFrame(category_list, columns=['id', 'category'])
category_df
```

```python
df_with_category = df.merge(category_df, left_on='category_id', right_on='id')
df_with_category[['title', 'category_id', 'category']]
```

```python
groupdatecate = df_with_category.groupby(['trending_date', 'category'])
groupsum = groupdatecate.views.sum().reset_index()
groupsum
```

```python
Sport_view=groupsum[groupsum.category == 'Sports'][['trending_date', 'views']]
Sport_view.rename(columns={'views': 'sport_views'}, inplace=True)
Sport_view
```

```python
Comedy_view=groupsum[groupsum.category == 'Comedy'][['trending_date', 'views']]
Comedy_view.rename(columns={'views': 'comedy_views'}, inplace=True)
Comedy_view
```

```python
Summary_view = Sport_view.merge(Comedy_view)
Summary_view
```

```python
condition_list = Summary_view['sport_views'] > Summary_view['comedy_views']
print(str(Summary_view[condition_list].shape[0])+" days")
```
