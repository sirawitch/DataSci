```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('https://github.com/kaopanboonyuen/Python-Data-Science/raw/master/Dataset/hed2020_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#Q1
print("Na on gill-size : "+str(df['gill-size'].isna().sum()))
```

```python
#drop label missing
df.dropna(subset=['label'],inplace=True)
```

```python
df
```

```python
#drop variable
df.drop(columns=['id','gill-attachment', 'gill-spacing', 'gill-size','gill-color-rate',
'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring-rate',
'stalk-color-below-ring-rate','veil-color-rate','veil-type'],inplace=True)
df
```

```python
#Q2
print("shape of data"+str(df.shape))
```

```python
#fill missing
fillmeanDf= df.fillna(df.mean())['cap-color-rate']
df.drop(columns=['cap-color-rate'],inplace=True)
fillmodeDf = df.fillna(df.mode().iloc[0])
newDf = pd.concat([fillmeanDf,fillmodeDf],axis=1)
newDf
```

```python
#Change label from nominal to number
dummieDf = pd.get_dummies(newDf[['label']])
dummieDf.drop(columns=['label_p'],inplace=True)
newDf.drop(columns=['label'],inplace=True)
df =  pd.concat([newDf,dummieDf],axis=1)
df
```

```python
#count label class0:1
numof0 = df[df.label_e == 0].count()["label_e"]
numof1 = df[df.label_e == 1].count()["label_e"]
print("number of 0,1 ="+str(numof0)+","+str(numof1))
```

```python
#convert nominal to numeric
df=pd.get_dummies(df,drop_first=True)
df
```

```python
#split train test
from sklearn.model_selection import train_test_split
y=df.pop('label_e')
X=df
X_train,X_test,y_train,y_test =train_test_split(X,y,stratify=y,test_size=0.2,random_state=2020)
```

```python
#Train and test row
print(X_train.shape[0])
print(X_test.shape[0])
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
rfc=RandomForestClassifier(random_state=2020)
param_grid={'criterion':['gini','entropy'],
'max_depth': [2,3,6],
'min_samples_leaf':[2,5,10],
'n_estimators':[100,200]}
CV_rfc=GridSearchCV(estimator=rfc,param_grid=param_grid,cv=5)
```

```python
CV_rfc.fit(X_train,y_train)
```

```python
#Q5 Best param
CV_rfc.best_params_
```

```python
rfc= CV_rfc.best_estimator_
```

```python
y_pred=rfc.predict(X_test)
```

```python
y_pred
```

```python
#Q6
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
cfm = confusion_matrix(y_true=y_test,y_pred=y_pred)
cfr = classification_report(y_true=y_test,y_pred=y_pred,target_names=['edible','poisoned'],digits=4)
print("Confusion matrix")
print(cfm)
print("Classification Report")
print(cfr)
```

```python
print("Macro f1")
classification_report(y_test,y_pred,digits=4,output_dict=True)['macro avg']['f1-score']
```
